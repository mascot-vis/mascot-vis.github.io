```python
import plotly.express as px
```

```python
df = px.data.gapminder().query("year == 2007")
```

```python
df
```

```python
df.to_csv("gapminder.csv")  
```

```python
continents = {}
for index, row in df.iterrows():
    cont = row["continent"]
    if cont not in continents:
        continents[cont] = {"name": cont, "type": "continent", "children": []}
    ctry = {}
    ctry['name'] = row['country']
    ctry['type'] = "country"
    ctry['pop'] = row['pop']
    ctry['lifeExp'] = row['lifeExp']
    ctry['gdpPercap'] = row['gdpPercap']
    continents[cont]['children'].append(ctry)
    
treejson = {'name': 'gapminder', 'children': []}
for c in continents:
    continents[c]['pop'] = sum(obj['pop'] for obj in continents[c]['children'])
    continents[c]['lifeExp'] = sum(obj['lifeExp'] for obj in continents[c]['children'])/len(continents[c]['children'])
    continents[c]['gdpPercap'] = sum(obj['gdpPercap'] for obj in continents[c]['children'])/len(continents[c]['children'])
    treejson['children'].append(continents[c])
    treejson['pop'] = sum(obj['pop'] for obj in treejson['children'])
    treejson['lifeExp'] = sum(obj['lifeExp'] for obj in treejson['children'])/len(treejson['children'])
    treejson['gdpPercap'] = sum(obj['gdpPercap'] for obj in treejson['children'])/len(treejson['children'])

import json
with open('gapminder.json', 'w') as file:
    file.write(json.dumps(treejson))
```

```python
import os

# Directory path
directory = '/Users/leozcliu/Code/Mascot'

# List directory contents

excludedFolders = ['scenes', 'node_modules', 'spec', '.git']

excludedFiles = ['publish.sh', 'package-lock.json', 'sandbox.html', 'sandbox_demos.json', 'di.sh', 'scenes.zip']

def getContents(directory):
    contents = os.listdir(directory)
    children = []
    for filename in contents:
        if filename[0] == '.':
            continue
        filepath = os.path.join(directory, filename)
        if os.path.isfile(filepath):
            if filename in excludedFiles:
                continue
            file_size = os.path.getsize(filepath)
            children.append({"name": filename, "size": file_size, "filetype": os.path.splitext(filepath)[1]})
        else:
            if filename not in excludedFolders:
                c = getContents(filepath)
                children.append({"name": filename, "children": c, "filetype": "folder", "size": sum(obj['size'] for obj in c)})
    return children
                
root = {"name": "Mascot", "filetype": "folder", "children": getContents(directory)} 
root['size'] = sum(obj['size'] for obj in root['children'])

import json
with open('mascot_1.12.1.json', 'w') as file:
    file.write(json.dumps(root))
```
